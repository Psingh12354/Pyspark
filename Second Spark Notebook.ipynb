```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
                    StructField("age", IntegerType(), True),
                    StructField("gender", StringType(), True),
                    StructField("name", StringType(), True),
                    StructField("course", StringType(), True),
                    StructField("roll", StringType(), True),
                    StructField("marks", IntegerType(), True),
                    StructField("email", StringType(), True)
])
df = spark.read.options(header='True').schema(schema).csv('/FileStore/tables/StudentData.csv')
df.show()
df.printSchema()
```

```
# Rdd to dataframe
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("RDD")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/StudentData.csv')
print(rdd.first())
headers = rdd.first()
print(type(headers))
print(headers)
rdd = rdd.filter(lambda x: x != headers).map(lambda x: x.split(','))
```

```
columns = headers.split(',')
dfRdd = rdd.toDF(columns)
dfRdd.show()
```

```
# another method of creating df from rdd
schema = StructType([
                    StructField("age", IntegerType(), True),
                    StructField("gender", StringType(), True),
                    StructField("name", StringType(), True),
                    StructField("course", StringType(), True),
                    StructField("roll", StringType(), True),
                    StructField("marks", IntegerType(), True),
                    StructField("email", StringType(), True)
])
print(schema)
```

```
# another method of creating df from rdd
rdd = rdd.map(lambda x: [int(x[0]), x[1], x[2], x[3], x[4], int(x[5]), x[6]])
dfRdd = spark.createDataFrame(rdd, schema=schema)
dfRdd.show()
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
```

```
df = spark.read.options(header='True', inferSchema='True').csv('/FileStore/tables/StudentData.csv')
df.show()
df.printSchema()
```

```
# changing column in one step if name not there create new else update 
# create a new rdd otherwise it become difficult to rollback as no option is available
df1 = df.withColumn("roll", col("roll").cast("String"))
```

```
#  create a new rdd otherwise it become difficult to rollback as no option is available
```

```
df = df.withColumn("marks", col('marks') + 10)
df.show()
```

```
df = df.withColumn("aggregated marks", col('marks') - 10)
df.show()
```

```
df = df.withColumn("name", lit("USA"))
df.show()
```

```
df = df.withColumn("marks", col("marks") - 10).withColumn("updated marks", col("marks") + 20).withColumn("Country", lit("USA"))
df.show()
```

```
# renaming column with entire structure
df = df.withColumnRenamed("gender","sex")
df.show()
df = df.withColumnRenamed("gender","sex").withColumnRenamed("roll", "roll number")
df.show()
```

```
# return only one column with new name it will not affect the actual data
df.select(col("name").alias("Full Name")).show()
df.show()
```

```
df.filter(df.course == "DB").show()
# or
df.filter(col("course") == "DB").show()
```

```
df.filter( (df.course == "DB") & (df.marks > 50) ).show()
```

```
courses = ["DB", "Cloud", "OOP", "DSA"]
df.filter(df.course.isin(courses)).show()
```

```
#startwith
df.filter(df.course.startswith("DS")).show()
#endswith
df.filter(df.course.endswith("d")).show()
```

```
# contains
df.filter(df.email.contains("Sheryll")).show()
```

```
# like is use to find similarity like any think before o and anythink after o here _ for one char and % for more than one char
df.filter(df.course.like('%_ou%')).show()
```

```
df.filter(df.email.like('S_nt%')).show()
```

```
df.filter(df.course=="DB").count()
```

```
df.select("course").distinct().count()
# inside show you can enter the number till which row do wanna show
df.select("course").distinct().show()
```

```
df1 = df.select("roll number").distinct().count()
counting = df.count()
if df1 == counting : 
    print("roll is unique")
else:
    print("Not unique")
```

```
# sorting top 100
df1 = df.withColumnRenamed("roll number","roll")
df1.sort(df1.marks).show(100)
```

```
# sort according to one or more column
df1.orderBy(df1.roll, df1.age).show()
```

```
# ascending and descending
df.sort(df.marks.asc(), df.age.desc()).show()
```

```
# SORT BY sorts data inside partition, while ORDER BY is global sort.

# SORT BY calls sortWithinPartitions() function, while ORDER BY calls sort()

# Both of these functions call sortInternal(), but with different global flag:
# def sortWithinPartitions ...
# sortInternal(global = false, sortExprs)

 

# def sort ...
# sortInternal(global = true, sortExprs)

# OrderBy is just an alias for the Sort function and should give the same result.


# https://towardsdatascience.com/sort-vs-orderby-in-spark-8a912475390
```
