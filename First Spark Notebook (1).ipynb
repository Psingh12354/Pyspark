```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Project")
sc = SparkContext.getOrCreate(conf=conf)
text = sc.textFile('/FileStore/tables/StudentData.csv')
```

```
header = text.first()
rdd=text.filter(lambda x: x!=header)
rdd=rdd.map(lambda x: x.split(','))
rdd.collect()
```

```
#count number of student
rdd.count()
```

```
# count marks sum
rdd2 = rdd
rdd2 =rdd2.map(lambda x : (x[1],int(x[5])))
rdd2.reduceByKey(lambda x,y: x+y).collect()
```

```
# passed or failed
rdd3 = rdd
passed = rdd3.filter(lambda x: int(x[5])>50).count()
failed = rdd3.filter(lambda x: int(x[5])<=50).count()
print(f'Passed : {passed}\nFailed : {failed}')
```

```
# total number of student enrolled per course
rdd4 = rdd
rdd4 = rdd.map(lambda x: (x[3],1))
rdd4 = rdd4.reduceByKey(lambda x,y : x+y)
rdd4.collect()
```

```
# total number of marks student achived per course
rdd5 = rdd
rdd5 = rdd.map(lambda x: (x[3],int(x[5])))
rdd5 = rdd5.reduceByKey(lambda x,y : x+y)
rdd5.collect()
```

```
# avg marks achive per course
rdd6 = rdd
rdd6 = rdd6.map(lambda x: (x[3], (int(x[5]), 1) ))
rdd6 = rdd6.reduceByKey( lambda x,y : (x[0] + y[0], x[1] + y[1]))
rdd6.map(lambda x: (x[0], (x[1][0]/x[1][1]))).collect()
```

```
# minimum and maximum marks per subject
rdd7 = rdd
rdd7 = rdd7.map(lambda x: (x[3], int(x[5])))
minimum = rdd7.reduceByKey(lambda x,y : x if x<y else y).collect()
maximum = rdd7.reduceByKey(lambda x,y : x if x>y else y).collect()
print(f'Minimum : {minimum}\nMaximum : {maximum}')
```

```
# Average age of male and female
rdd8 = rdd
rdd8 = rdd8.map(lambda x: (x[1], (int(x[0]), 1) ))
rdd8 = rdd8.reduceByKey( lambda x,y : (x[0] + y[0], x[1] + y[1]))
rdd8.map(lambda x: (x[0], (x[1][0]/x[1][1]))).collect()
```
