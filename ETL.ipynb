```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('ETL').getOrCreate()
```

```
df=spark.read.text('/FileStore/tables/autobiography.txt')
```

```
df.show()
```

```
df.collect()
```

```
display(df)
```

```
from pyspark.sql.functions import lit,col,explode
import pyspark.sql.functions as f
```

```
df2 = df.withColumn("SplitData",f.split('value',' '))
display(df2)
```

```
# to split the splitted data in multiple line use explode
df3 = df2.withColumn('word',explode('SplitData'))
wordDf = df3.select("word")
wordDf.show()
```

```
wordDf.groupBy('word').count().show()
```
