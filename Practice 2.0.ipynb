```
from pyspark.sql import SparkSession
from pyspark.sql.functions import count,sum,avg,min,mean,max
```

```
spark = SparkSession.builder.appName("Office").getOrCreate()
```

```
df = spark.read.options(inferSchema=True,header=True,delimeter=',').csv('/FileStore/tables/StudentData.csv')
df.show()
```

```
df.printSchema()
```

```
df.orderBy('bonus').show(10)
# or
df.sort(df.bonus.asc()).show(10)
```

```
df.sort(df.age.desc(),df.salary.asc()).show()
```

```
df.sort(df.age.desc(),df.bonus.desc(),df.salary.asc()).show()
```

```
# to perform groupBy you must perform someother opertation to on it else it will give an error
```

```
df.groupBy("department","state").count().show()
```

```
df.withColumn("Inhand Salary", ((df.salary - df.bonus)/13)*12).show()
```

```
df.groupBy('department').agg(count('*').alias('count'),sum('salary').alias('salary'),sum('bonus').alias("bonus")).show()
```

```
df.groupBy('department').sum().show()
# or 
df.groupBy('department').sum('salary').show()
```

```
df.groupBy('gender').count().show()
```

```
df.filter(df.marks>50).groupBy('gender').count().show()
```

```
df1=df.filter(df.gender=='Female')
df1.show(5)
```

```
df2 = df1.groupBy(df1.course,df1.name).agg(count('*'))
df2.show()
```

```
df.groupBy(df.course).count().show()
```

```
df.groupBy(df.course,df.gender).count().show()
```

```
df.groupBy(df.course,df.gender).sum('marks').show()
```

```
df.groupBy(df.course,df.age).agg(min('marks'),max('marks'),avg('marks')).show()
```
