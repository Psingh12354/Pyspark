```
import pyspark
from pyspark import SparkContext,SparkConf
conf = SparkConf().setAppName("demoprogram")
sc = SparkContext.getOrCreate(conf=conf)
text = sc.textFile("/FileStore/tables/filter.txt")
text.collect()
```

```
splitwords = text.flatMap(lambda x: x.split(" "))
splitwords.collect()
```

```
# Startwith return true if element is present
def AandC(x):
    if x.startswith('a') or x.startswith('c'):
        return False
    else:
        return True
```

```
filterRdd= splitwords.filter(AandC)
filterRdd.collect()
```

```
filterRdd= splitwords.filter(lambda x: not(x.startswith('a') or x.startswith('c')))
filterRdd.collect()
```

```
#avg
import pyspark
from pyspark import SparkContext,SparkConf
conf = SparkConf().setAppName("reduce")
sc = SparkContext.getOrCreate(conf=conf)
text = sc.textFile("/FileStore/tables/movie.txt")
text.collect()
```

```
ratings = sc.textFile("/FileStore/tables/movie.txt")
ratings.collect()
```

```
rdd2 = ratings.map(lambda x: (x.split(",")[0],(int(x.split(",")[1]),1)))
rdd2.collect()
```

```
rdd3 = rdd2.reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1]))
rdd3.collect()
```

```
rdd4 = rdd3.map(lambda x: (x[0],x[1][0]/x[1][1]))
rdd4.collect()
```
